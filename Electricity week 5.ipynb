```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
y = [5.13579843705026,
4.86753445045558,
4.60517018598809,
4.78749174278205,
4.24849524204936,
3.68887945411394,
3.91202300542815,
3.91202300542815,
3.91202300542815,
3.40119738166216,
2.99573227355399,
2.99573227355399]
x = [1.90954250488444,
1.83258146374831,
1.74919985480926,
1.65822807660353,
1.55603713570699,
1.44691898293633,
1.32175583998232,
1.17865499634165,
1.01160091167848,
0.810930216216329,
0.559615787935423,
0.22314355131421]

par = np.polyfit(x,y,1,full='True')
slope=par[0][0]

intercept=par[0][1]

variance = np.var(y)
a = [min(x), max(x)]
b = [min(x) * slope + intercept, max(x) * slope + intercept]

plt.scatter(x,y,marker='+')
plt.plot(a,b)
plt.xlabel('ln(rmid)')
plt.ylabel('ln(E)')
plt.title('ln(E) vs ln(rmid)')
plt.show()
print('Best fit line - y = %3fx + %3f' % (slope, intercept))
print('variance -', variance)
```
